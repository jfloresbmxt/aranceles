```python
import pandas as pd
import numpy as np
import os

# Ajusta tu ruta si es necesario
# os.chdir('C:/Users/M20537/Downloads')
csv_path = "htsdata.csv"
nombre_salida = "HTS_Completa_2025.xlsx"

# ==========================================
# 1. DICCIONARIOS Y CONFIGURACIÓN
# ==========================================

SECTIONS_NAMES_EN = {
    "I": "LIVE ANIMALS; ANIMAL PRODUCTS",
    "II": "VEGETABLE PRODUCTS",
    "III": "ANIMAL OR VEGETABLE FATS AND OILS...",
    "IV": "PREPARED FOODSTUFFS; BEVERAGES, SPIRITS AND VINEGAR...",
    "V": "MINERAL PRODUCTS",
    "VI": "PRODUCTS OF THE CHEMICAL OR ALLIED INDUSTRIES",
    "VII": "PLASTICS AND ARTICLES THEREOF; RUBBER AND ARTICLES THEREOF",
    "VIII": "RAW HIDES AND SKINS, LEATHER, FURSKINS...",
    "IX": "WOOD AND ARTICLES OF WOOD; WOOD CHARCOAL...",
    "X": "PULP OF WOOD OR OF OTHER FIBROUS CELLULOSIC MATERIAL...",
    "XI": "TEXTILES AND TEXTILE ARTICLES",
    "XII": "FOOTWEAR, HEADGEAR, UMBRELLAS, SUN UMBRELLAS...",
    "XIII": "ARTICLES OF STONE, PLASTER, CEMENT, ASBESTOS, MICA...",
    "XIV": "NATURAL OR CULTURED PEARLS, PRECIOUS OR SEMI-PRECIOUS STONES...",
    "XV": "BASE METALS AND ARTICLES OF BASE METAL",
    "XVI": "MACHINERY AND MECHANICAL APPLIANCES; ELECTRICAL EQUIPMENT...",
    "XVII": "VEHICLES, AIRCRAFT, VESSELS AND ASSOCIATED TRANSPORT EQUIPMENT",
    "XVIII": "OPTICAL, PHOTOGRAPHIC, CINEMATOGRAPHIC, MEASURING...",
    "XIX": "ARMS AND AMMUNITION; PARTS AND ACCESSORIES THEREOF",
    "XX": "MISCELLANEOUS MANUFACTURED ARTICLES",
    "XXI": "WORKS OF ART, COLLECTORS' PIECES AND ANTIQUES",
    "XXII": "SPECIAL CLASSIFICATION PROVISIONS"
}

CHAPTERS_NAMES_EN = {
    "01": "Live animals", "02": "Meat and edible meat offal",
    "03": "Fish and crustaceans, molluscs and other aquatic invertebrates",
    "04": "Dairy produce; birds' eggs; natural honey...",
    "05": "Products of animal origin, not elsewhere specified or included",
    "06": "Live trees and other plants; bulbs, roots and the like...",
    "07": "Edible vegetables and certain roots and tubers",
    "08": "Edible fruit and nuts; peel of citrus fruit or melons",
    "09": "Coffee, tea, maté and spices",
    "10": "Cereals",
    "11": "Products of the milling industry; malt; starches...",
    "12": "Oil seeds and oleaginous fruits; miscellaneous grains...",
    "13": "Lac; gums, resins and other vegetable saps and extracts",
    "14": "Vegetable plaiting materials; vegetable products not elsewhere specified",
    "15": "Animal or vegetable fats and oils...",
    "16": "Preparations of meat, of fish or of crustaceans...",
    "17": "Sugars and sugar confectionery",
    "18": "Cocoa and cocoa preparations",
    "19": "Preparations of cereals, flour, starch or milk...",
    "20": "Preparations of vegetables, fruit, nuts or other parts of plants",
    "21": "Miscellaneous edible preparations",
    "22": "Beverages, spirits and vinegar",
    "23": "Residues and waste from the food industries...",
    "24": "Tobacco and manufactured tobacco substitutes",
    "25": "Salt; sulphur; earths and stone; plastering materials...",
    "26": "Ores, slag and ash",
    "27": "Mineral fuels, mineral oils and products of their distillation...",
    "28": "Inorganic chemicals; organic or inorganic compounds...",
    "29": "Organic chemicals", "30": "Pharmaceutical products", "31": "Fertilisers",
    "32": "Tanning or dyeing extracts; tannins and their derivatives...",
    "33": "Essential oils and resinoids; perfumery, cosmetic or toilet preparations",
    "34": "Soap, organic surface-active agents, washing preparations...",
    "35": "Albuminoidal substances; modified starches; glues; enzymes",
    "36": "Explosives; pyrotechnic products; matches...",
    "37": "Photographic or cinematographic goods",
    "38": "Miscellaneous chemical products",
    "39": "Plastics and articles thereof", "40": "Rubber and articles thereof",
    "41": "Raw hides and skins (other than furskins) and leather",
    "42": "Articles of leather; saddlery and harness; travel goods...",
    "43": "Furskins and artificial fur; manufactures thereof",
    "44": "Wood and articles of wood; wood charcoal",
    "45": "Cork and articles of cork",
    "46": "Manufactures of straw, of esparto or of other plaiting materials...",
    "47": "Pulp of wood or of other fibrous cellulosic material...",
    "48": "Paper and paperboard; articles of paper pulp...",
    "49": "Printed books, newspapers, pictures and other products...",
    "50": "Silk", "51": "Wool, fine or coarse animal hair...", "52": "Cotton",
    "53": "Other vegetable textile fibers; paper yarn...",
    "54": "Man-made filaments; strip and the like...",
    "55": "Man-made staple fibers",
    "56": "Wadding, felt and nonwovens; special yarns...",
    "57": "Carpets and other textile floor coverings",
    "58": "Special woven fabrics; tufted textile fabrics...",
    "59": "Impregnated, coated, covered or laminated textile fabrics...",
    "60": "Knitted or crocheted fabrics",
    "61": "Articles of apparel and clothing accessories, knitted or crocheted",
    "62": "Articles of apparel and clothing accessories, not knitted or crocheted",
    "63": "Other made up textile articles; sets; worn clothing...",
    "64": "Footwear, gaiters and the like; parts of such articles",
    "65": "Headgear and parts thereof",
    "66": "Umbrellas, sun umbrellas, walking-sticks, seat-sticks...",
    "67": "Prepared feathers and down and articles made of feathers...",
    "68": "Articles of stone, plaster, cement, asbestos, mica...",
    "69": "Ceramic products", "70": "Glass and glassware",
    "71": "Natural or cultured pearls, precious or semi-precious stones...",
    "72": "Iron and steel", "73": "Articles of iron or steel",
    "74": "Copper and articles thereof", "75": "Nickel and articles thereof",
    "76": "Aluminum and articles thereof",
    "77": "(Reserved for possible future use in the Harmonized System)",
    "78": "Lead and articles thereof", "79": "Zinc and articles thereof",
    "80": "Tin and articles thereof",
    "81": "Other base metals; cermets; articles thereof",
    "82": "Tools, implements, cutlery, spoons and forks...",
    "83": "Miscellaneous articles of base metal",
    "84": "Nuclear reactors, boilers, machinery and mechanical appliances...",
    "85": "Electrical machinery and equipment and parts thereof...",
    "86": "Railway or tramway locomotives, rolling-stock and parts thereof...",
    "87": "Vehicles other than railway or tramway rolling-stock...",
    "88": "Aircraft, spacecraft, and parts thereof",
    "89": "Ships, boats and floating structures",
    "90": "Optical, photographic, cinematographic, measuring...",
    "91": "Clocks and watches and parts thereof",
    "92": "Musical instruments; parts and accessories of such articles",
    "93": "Arms and ammunition; parts and accessories thereof",
    "94": "Furniture; bedding, mattresses, mattress supports...",
    "95": "Toys, games and sports requisites; parts and accessories thereof",
    "96": "Miscellaneous manufactured articles",
    "97": "Works of art, collectors' pieces and antiques",
    "98": "Special classification provisions"
}

RANGOS_SECCIONES = [
    (1, 5, "I"), (6, 14, "II"), (15, 15, "III"), (16, 24, "IV"),
    (25, 27, "V"), (28, 38, "VI"), (39, 40, "VII"), (41, 43, "VIII"),
    (44, 46, "IX"), (47, 49, "X"), (50, 63, "XI"), (64, 67, "XII"),
    (68, 70, "XIII"), (71, 71, "XIV"), (72, 83, "XV"), (84, 85, "XVI"),
    (86, 89, "XVII"), (90, 92, "XVIII"), (93, 93, "XIX"), (94, 96, "XX"),
    (97, 97, "XXI"), (98, 98, "XXII")
]
MAPA_CAPITULO_SECCION = {}
for inicio, fin, romana in RANGOS_SECCIONES:
    for i in range(inicio, fin + 1):
        MAPA_CAPITULO_SECCION[f"{i:02d}"] = romana

def get_seccion_romana(cap_code):
    return MAPA_CAPITULO_SECCION.get(cap_code, "ND")

# ==========================================
# 2. CARGA Y LIMPIEZA INICIAL
# ==========================================
print("Cargando archivo CSV...")
df = pd.read_csv(csv_path, dtype=str)

# Quitar ultimas 3 columnas
df = df.iloc[:, :-3]

df = df.rename(columns={
    'HTS Number': 'HTS_Number',
    'Indent': 'Indent',
    'Description': 'Description',
    'Unit of Quantity': 'Unit',
    'General Rate of Duty': 'General_Rate',
    'Special Rate of Duty': 'Special_Rate'
})

# Limpieza de NaNs al inicio
df = df.fillna('')

df['Indent'] = pd.to_numeric(df['Indent'], errors='coerce').fillna(0).astype(int)

print("Limpiando puntos y cortando en Capítulo 99...")
df['HTS_Number'] = df['HTS_Number'].str.replace('.', '', regex=False).str.strip()

# --- MODIFICACIÓN: CORTAR EN EL PRIMER 99 (y descartar todo lo de abajo) ---
# Buscamos los índices donde el código empieza con '99'
indices_cap99 = df.index[df['HTS_Number'].str.startswith('99', na=False)].tolist()

if indices_cap99:
    # Tomamos el primer índice encontrado (el primer 99 de arriba hacia abajo)
    corte_idx = indices_cap99[0]
    print(f"Capítulo 99 detectado en la fila {corte_idx}. Cortando datos...")
    # Recortamos el DataFrame hasta antes de ese índice
    df = df.iloc[:corte_idx]
else:
    print("No se detectó Capítulo 99. Se procesa todo el archivo.")

# ==========================================
# 3. LÓGICA DE TRANSFORMACIÓN DE CÓDIGOS
# ==========================================
print("Aplicando reglas de Indentación (Hijo llena a Padre)...")

hts_codes = df['HTS_Number'].tolist()
indents = df['Indent'].tolist()
n = len(df)

for i in range(n):
    current_indent = indents[i]
    current_code = str(hts_codes[i]) if pd.notna(hts_codes[i]) else ""

    if current_indent == 1:
        if current_code == "" or current_code.lower() == "nan":
            if i + 1 < n:
                next_indent = indents[i+1]
                next_code = str(hts_codes[i+1])
                if next_indent == 2 and next_code != "" and next_code.lower() != "nan":
                    fill_val = next_code[:5]
                    hts_codes[i] = fill_val
                    current_code = fill_val

        if len(current_code) > 5:
            hts_codes[i] = current_code[:5]

    elif current_indent == 0:
        if len(current_code) > 4:
            hts_codes[i] = current_code[:4]

df['HTS_Number'] = hts_codes

# --- ORIGINAL CLEANED (Con puntos) ---
df_original_cleaned = df.copy()

def restore_dots_original(code):
    s = str(code)
    if len(s) <= 4: return s
    elif len(s) <= 6: return f"{s[:4]}.{s[4:]}"
    elif len(s) <= 8: return f"{s[:4]}.{s[4:6]}.{s[6:]}"
    else: return f"{s[:4]}.{s[4:6]}.{s[6:8]}.{s[8:]}"

df_original_cleaned['HTS_Number'] = df_original_cleaned['HTS_Number'].apply(restore_dots_original)

# ==========================================
# 4. CREACIÓN DE DICCIONARIOS
# ==========================================
print("Generando diccionarios de descripciones...")

dict_partidas = {}
dict_desdoblamientos = {}
dict_subpartidas = {}
dict_items = {}

for row in df.itertuples():
    code = str(row.HTS_Number)
    desc = str(row.Description)
    if code == '': continue

    length = len(code)
    if length == 4: dict_partidas[code] = desc
    elif length == 5: dict_desdoblamientos[code] = desc
    elif length == 6: dict_subpartidas[code] = desc
    elif length == 8: dict_items[code] = desc

# ==========================================
# 5. FILTRADO PARA PANELES
# ==========================================
print("Filtrando filas con Arancel para paneles analíticos...")

df_final = df[df['General_Rate'].notna() & (df['General_Rate'] != '')].copy()

print("Normalizando códigos para paneles (Padding y Truncado)...")

def normalize_to_8(code):
    s = str(code)
    if len(s) > 8: s = s[:8]
    return s.ljust(8, '0')

df_final['HTS_Number'] = df_final['HTS_Number'].apply(normalize_to_8)

# ==========================================
# 6. CONSTRUCCIÓN DE PANELES
# ==========================================
print("Construyendo columnas del panel...")

panel_df = df_final.copy()
panel_df['Codigo_Limpio'] = panel_df['HTS_Number']

panel_df['Seccion_Romana'] = panel_df['Codigo_Limpio'].str[:2].apply(get_seccion_romana)
panel_df['Seccion_Nombre'] = panel_df['Seccion_Romana'].map(SECTIONS_NAMES_EN)

panel_df['Capitulo_Num'] = panel_df['Codigo_Limpio'].str[:2]
panel_df['Capitulo_Nom'] = panel_df['Capitulo_Num'].map(CHAPTERS_NAMES_EN)

panel_df['Partida_Num'] = panel_df['Codigo_Limpio'].str[2:4]
panel_df['Partida_Key'] = panel_df['Codigo_Limpio'].str[:4]
panel_df['Partida_Nom'] = panel_df['Partida_Key'].map(dict_partidas)

panel_df['Desdoblamiento_Num'] = panel_df['Codigo_Limpio'].str[4:5]
panel_df['Desdoblamiento_Key'] = panel_df['Codigo_Limpio'].str[:5]
panel_df['Desdoblamiento_Nom'] = panel_df['Desdoblamiento_Key'].map(dict_desdoblamientos)

panel_df['Subpartida_Num'] = panel_df['Codigo_Limpio'].str[5:6]
panel_df['Subpartida_Key'] = panel_df['Codigo_Limpio'].str[:6]
panel_df['Subpartida_Nom'] = panel_df['Subpartida_Key'].map(dict_subpartidas)

panel_df['Fraccion_Num'] = panel_df['Codigo_Limpio'].str[6:8]

def format_hts_8(code):
    return f"{code[:4]}.{code[4:6]}.{code[6:8]}"

panel_df['Codigo_Formato'] = panel_df['Codigo_Limpio'].apply(format_hts_8)

# ==========================================
# 7. DEFINICIÓN DE COLUMNAS DE SALIDA
# ==========================================

cols_num = ['Codigo_Formato', 'Seccion_Romana', 'Capitulo_Num', 'Partida_Num', 'Desdoblamiento_Num', 'Subpartida_Num', 'Fraccion_Num', 'Unit', 'General_Rate', 'Special_Rate']
df_numerico = panel_df[cols_num].copy()
df_numerico.columns = ['Code', 'Section', 'Chapter', 'Heading', 'Breakdown', 'Subheading', 'Item', 'Unit', 'General Duty', 'Special Duty']

cols_txt = ['Codigo_Formato', 'Seccion_Nombre', 'Capitulo_Nom', 'Partida_Nom', 'Desdoblamiento_Nom', 'Subpartida_Nom', 'Description', 'Unit', 'General_Rate', 'Special_Rate']
df_textual = panel_df[cols_txt].copy()
df_textual.columns = ['Code', 'Section Name', 'Chapter Name', 'Heading Name', 'Breakdown Name', 'Subheading Name', 'Description', 'Unit', 'General Duty', 'Special Duty']

cols_ext = ['Codigo_Formato', 'Seccion_Romana', 'Seccion_Nombre', 'Capitulo_Num', 'Capitulo_Nom', 'Partida_Num', 'Partida_Nom', 'Desdoblamiento_Num', 'Desdoblamiento_Nom', 'Subpartida_Num', 'Subpartida_Nom', 'Fraccion_Num', 'Description', 'Unit', 'General_Rate', 'Special_Rate']
df_extendido = panel_df[cols_ext].copy()
df_extendido.columns = ['Code', 'Section', 'Section Name', 'Chapter', 'Chapter Name', 'Heading', 'Heading Name', 'Breakdown', 'Breakdown Name', 'Subheading', 'Subheading Name', 'Item', 'Description', 'Unit', 'General Duty', 'Special Duty']

# Limpiezas finales
df_numerico = df_numerico.fillna('')
df_textual = df_textual.fillna('')
df_extendido = df_extendido.fillna('')
df_original_cleaned = df_original_cleaned.fillna('')

# ==========================================
# 8. EXPORTACIÓN EXCEL
# ==========================================

print(f"Creando archivo Excel: {nombre_salida}...")
writer = pd.ExcelWriter(nombre_salida, engine='xlsxwriter')
workbook = writer.book

fmt_header = workbook.add_format({'bold': True, 'border': 1, 'bg_color': '#D9E1F2', 'valign': 'top', 'align': 'center', 'font_name': 'Arial', 'font_size': 10})
fmt_normal = workbook.add_format({'border': 1, 'valign': 'top', 'font_name': 'Arial', 'font_size': 9})
fmt_wrap = workbook.add_format({'border': 1, 'valign': 'top', 'text_wrap': True, 'font_name': 'Arial', 'font_size': 9})

def escribir_hoja(dataframe, nombre_hoja):
    worksheet = workbook.add_worksheet(nombre_hoja)
    writer.sheets[nombre_hoja] = worksheet
    headers = dataframe.columns.tolist()
    for col_idx, header in enumerate(headers):
        worksheet.write(0, col_idx, header, fmt_header)
    for row_idx, row in enumerate(dataframe.itertuples(index=False), start=1):
        for col_idx, value in enumerate(row):
            header_name = str(headers[col_idx])

            # Wrap solo para Nombres y Descripción (Special Duty NO)
            es_wrap_text = any(x in header_name for x in ["Name", "Description"])

            formato = fmt_wrap if es_wrap_text else fmt_normal
            if pd.isna(value): value = ""
            worksheet.write(row_idx, col_idx, value, formato)

    for idx, col_name in enumerate(headers):
        header_str = str(col_name)

        if any(x in header_str for x in ["Name", "Description"]):
            worksheet.set_column(idx, idx, 50)

        elif any(x in header_str for x in ["Special", "Special_Rate", "Special Duty"]):
            # Special Duty: Autofit (Tope 150)
            max_len = dataframe[col_name].astype(str).map(len).max() if not dataframe.empty else 0
            final_width = min(max(max_len, len(header_str)) + 2, 150)
            worksheet.set_column(idx, idx, final_width)

        else:
            # Resto: Autofit (Tope 25)
            max_len = dataframe[col_name].astype(str).map(len).max() if not dataframe.empty else 0
            final_width = min(max(max_len, len(header_str)) + 2, 25)
            worksheet.set_column(idx, idx, final_width)

print("Exportando Original Limpia (Completa)...")
escribir_hoja(df_original_cleaned, "Original Cleaned")

print("Exportando Panel Numérico...")
escribir_hoja(df_numerico, "Numeric Panel")

print("Exportando Panel Textual...")
escribir_hoja(df_textual, "Textual Panel")

print("Exportando Panel Extendido...")
escribir_hoja(df_extendido, "Extended Panel")

writer.close()
print("¡ÉXITO! Proceso HTS USA terminado.")
```

Vamos a trabajar con una base de datos llamada "htsdata". En esta base de datos tenemos los códigos arancelarios, y las tasas correspondientes. La idea es sencilla, vamos a utilizar un documento de la LIGIE para rellenar los huecos de la HTS.

Primero que nada tenemos que limpiar la base inicial, quedándonos únicamente con las columnas "HTS Number", "Description", "Unit of Quantity", "General Rate of Duty" y "Special Rate of Duty". A nivel de filas, quitaremos todo lo que venga posterior al primer código que empiece en 99 (De arriba a abajo). El data frame que resulte quedará en la primer hoja del excel que exportaremos, con formato de azul claro en los encabezados y negritas. Además, queremos que tenga bordes de tabla todo el contenido del mismo.

Luego, haremos 4 paneles, uno numérico, uno textual, uno extendido y uno concatenado.
Para esto, es primero necesario que nos quedémos únicamente con las filas que tienen contenido en la columna de "General Rate of Duty", posteriormente, de los códigos que "sobrevivan", vamos a quitarle los últimos dos dígitos a todos los códigos que tengan 10 números o 13 caracteres en caso de contemplar los puntos, recomendaría que desde el inicio se eliminarán los puntos de esta columna para un efectivo tratamiento.

En todos los paneles hay que mantener las columnas de Unit of Quantity" Y "General Rate of Duty".

Númerico: Vamos a extraer los números que conforman el código arancelario, los primeros dos son el capítulo, los siguientes dos son la Partida, el siguiente dígito es el "Desdoblamiento", el sexto es la subpartida. Cada uno de estos números irá en sus columnas correspondientes.

Textual: Aquí haremos consultas al documento "LIGIE_Maestra_Unificada", este documento tiene una estructura muy similar al que queremos obtener. La idea es, que las columnas correspondientes a los primeros 6 dígitos, Capítulo, Partida, Desdoblamiento y Subpartida sean rellenados con la LIGIE, y los últimos 2 dígitos, correspondan a la "Fracción" que corresponde a la columna de "Descripción".

Extendido: Básicamente significa combinar el Númerico y Textual.

Concatenado: Consiste en tener la columna de código, la concatenación de todas las columnas textuales, y las columnas de Unit of Quantity" Y "General Rate of Duty".
